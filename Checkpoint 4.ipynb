# question 3

```python
def addition(liste):  
    R = 0
    for i in liste:    
        R += i  
    return R  
print ("la somme de la liste_1 est",addition([1,2,2,6]))
 
def multiply(liste):  
    R = 1
    for y in liste:
        R *= y  
    return R  
print("la multiplication  de la liste_2 est",multiply((2, 2, 3)))
```

```python
l=[1,2,3,4]
pair=[]
impair= []
for x in range (len(l)):
    if x%2==0:
            pair.append(l[x])
    else :
            impair.append(l[x])
print ("la somme de la liste est",addition(pair))
print("la multiplication  de la liste est",multiply(impair))
```
